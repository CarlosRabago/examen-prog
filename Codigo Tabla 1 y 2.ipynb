```python
import pandas as pd

# se realiza lectura de datos del archivo .csv, se determina como indice la columna ID_REGISTRO

df = pd.read_csv("200511COVID19MEXICO.csv",index_col=1)

df.head()
```

```python
# Consulta a los encabezados de las columnas
df.columns
```

### Conﬁrmados de SARS-CoV2 en Sonora por fecha usando la fecha de inicio de síntomas (no acumulados)

```python
# Se condiciona la columan RESULTADO a que muestre solo valores "1" (Positivo SARS-CoV-2) y a la 
# columna ENTIDAD_RES muestre solo valores "26" (SONORA)

df2=df.loc[df["RESULTADO"] == 1&(df["ENTIDAD_RES"] == 26),
             ['ENTIDAD_RES','FECHA_INGRESO','FECHA_SINTOMAS', 'FECHA_DEF','RESULTADO']]

print(df2)
```

```python
# Se guardan la consulta anterior en el archivo "tablas1.csv"
df2.reset_index().to_csv("tablas1.csv", header=True, index=False)
```

### Generar una tabla con la cantidad de casos hospitalizados en los estados de Sonora, Chihuahua, Nuevo León y Puebla.

```python
# Se condiciona la columna ENTIDAD_RES a mostrar resultados [8,19,21,26] que corresponden a Chihuahua, 
# Nuevo Leon, Puebla, y Sonora, condicionado tambien por TIPO_PACIENTE sea igual a 2 (HOSPITALIZADO)

df3 = df.loc[df['ENTIDAD_RES'].isin([8,19,21,26]) & (df["TIPO_PACIENTE"] == 2),['ENTIDAD_RES',"TIPO_PACIENTE"]]
df3
```

```python
# Se guardan la consulta anterior en el archivo "tablas2.csv"
df3.reset_index().to_csv("tablas2.csv", header=True, index=False)
```
